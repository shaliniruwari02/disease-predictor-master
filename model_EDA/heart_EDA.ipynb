```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
df.describe().T
```

```python
f, ax = plt.subplots(figsize=(10, 8))
ax = sns.countplot(x="target", data=df)
plt.show()
```

```python
df.plot(kind='kde', subplots=True, layout=(7,2), sharex=False , figsize =(20,20))
plt.show()
```

```python
df.hist(figsize = (12,12), color = "#0504aa")
plt.show()
```

```python
df.plot(kind='box', subplots=True, layout=(7,2), figsize =(18,18))
plt.show()
```

```python
correlation = df.corr()
```

```python
correlation['target'].sort_values(ascending=False)
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x="target", y="thalach", data=df)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title('Correlation Heatmap of Heart Disease')
a = sns.heatmap(correlation, annot=True, square=True, linecolor='white')
a.set_xticklabels(a.get_xticklabels())
a.set_yticklabels(a.get_yticklabels())           
plt.show()
```

```python
sns.pairplot(df, kind='scatter')
plt.show()
```
